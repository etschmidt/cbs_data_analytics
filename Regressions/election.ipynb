```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
election = pd.read_csv('election_data.csv')
```

```python
election
```

```python
election['population'].plot.hist(bins=30)
```

```python
print(smf.ols('trump_share ~ population', data=election,).fit().summary())
```

```python
np.log(election['population']).plot.hist(bins=30)
```

```python
print(smf.ols('trump_share ~ np.log(population)', data=election,).fit().summary())
```

```python
print(smf.ols('trump_share ~ income', data=election,).fit().summary())
```

Coefficient for income significantly decreases when controlled for population

```python
smf.ols('trump_share ~ income', data=election,).fit().params['income']*10000
```

```python
smf.ols('trump_share ~ np.log(population) + income', data=election,).fit().params['income']*10000
```
