```python
ghr = pd.read_csv('C:/Users/bz2181/Desktop/ghr_sample100k.csv')
```

```python
ghr
```

```python
print(smf.ols('time_to_fill ~ salary',data=ghr).fit().summary())
```

```python
ghr['sal_sq'] = ghr['salary']**2
print(smf.ols('time_to_fill ~ salary + sal_sq',data=ghr).fit().summary())
```

```python
import numpy as np
```

```python
print(smf.ols('time_to_fill ~ np.log(salary)',data=ghr).fit().summary())
```
